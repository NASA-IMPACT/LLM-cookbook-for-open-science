```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def print_metrics(metrics,mode):
    print(f"The metrics {mode} finetuning is:")
    predictions=metrics
    aspect_metrics = {}
    for aspect, preds in predictions.items():
        accuracy = accuracy_score(preds, [1]*len(preds))  # Assuming ground truth is always 1
        aspect_metrics[aspect] = accuracy
    return aspect_metrics
```

```python
before_metrics=print_metrics(before_metrics,"25")
print(before_metrics)
```

```python
finetuning_10=print_metrics(finetuning10_metrics,"25")
print(finetuning_10)
```

```python
finetuning_5=print_metrics(finetuning5_metrics,"25")
print(finetuning_5)
```

```python
import matplotlib.pyplot as plt

# Extracting keys and values
keys = list(before_metrics.keys())
values_dict1 = list(before_metrics.values())
values_dict2 = list(finetuning_10.values())
values_dict3 = list(finetuning_5.values())

# Plotting
plt.figure(figsize=(10, 6))
bar_width = 0.2
index = range(len(keys))

plt.bar(index, values_dict1, width=bar_width, label='Before Finetuning', color='b', align='center')
plt.bar([i + bar_width for i in index], values_dict2, width=bar_width, label='Finetuned (Epoch=10)', color='g', align='center')
plt.bar([i + 2 * bar_width for i in index], values_dict3, width=bar_width, label='Finetuned (Epoch=5)', color='r', align='center')

plt.xlabel('Features')
plt.ylabel('Accuracy')
plt.title('Human evaluated values for each Key in various before and after finetuning steps for 25 datapoints')
plt.xticks([i + bar_width for i in index], keys, rotation=45)
plt.legend()

plt.tight_layout()
plt.show()
```
